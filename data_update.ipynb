```python
#url = "https://www.sportsbettingdime.com/soccer/world-cup-odds/"


url = "https://euenergy.live/"
```

```python
import pandas as pd

tables = pd.read_html(url)
len(tables)
```

```python
# Find the table I am interested in
df = tables[0]
df
```

```python
df = df.rename(columns={"Unnamed: 0": "country", "Unnamed: 1": "development", "â‚¬/MWh": "price"})
df
```

```python
df = df.astype(str).apply(lambda x: x.str.encode('ascii', 'ignore').str.decode('ascii'))
```

```python
df['country'] = df['country'].str.replace('Czechia', 'Czech Rep.')
```

```python
df['price'] = df['price'].astype(float)

#df = df.groupby('Country').Price.mean()
#df
```

```python
df = round(df.groupby('country').price.mean())
```

```python
df.to_csv("euenergy.csv")
```
