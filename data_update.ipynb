# Data update script for ....



```python
url = "https://en.wikipedia.org/wiki/List_of_most_expensive_paintings#List_of_highest_prices_paid"
```

```python
import pandas as pd

tables = pd.read_html(url)
len(tables)
```

```python
# Find the table I am interested in
df = tables[0]
df
```

```python
df.dtypes
updated_df = df['Auction house'].value_counts().to_frame(name='count').reset_index
updated_df
```

```python
updated_df.to_csv('auction_house.csv', index=False)
```
