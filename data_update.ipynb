# Data update script for ....



```python
import pandas as pd

tables = pd.read_html("https://euenergy.live/")
len(tables)
```

```python
# Find the table I am interested in
df = tables[0]
df
```

```python
#Rename the columns
df = df.rename(columns={"Unnamed: 0": "country", "Unnamed: 1": "development", "€/MWh": "price"})
df
```

```python
#Remove the emojis
df = df.astype(str).apply(lambda x: x.str.encode('ascii', 'ignore').str.decode('ascii'))
```

```python
# And rename the Czech Republic to match datawrapper
df['country'] = df['country'].str.replace('Czechia', 'Czech Rep.')
```

```python
# Turn the price column into a float
df['price'] = df['price'].astype(float)
```

```python
# Group the data by country and find the mean price per country to get the prices on country leegl
df = round(df.groupby('country').price.mean())
```

```python
df.to_csv("europaenergy.csv")
```
